```python
import pandas as pd
cancerdf = pd.read_csv('/content/paad_tcga_pan_can_atlas_2018_clinical_data.tsv', sep='\t')

a = cancerdf.groupby(['Neoplasm Histologic Grade',
                      'Neoplasm Disease Stage American Joint Committee on Cancer Code',
                      'Radiation Therapy'] ).agg(
    Survival_months_mean=('Overall Survival (Months)', 'mean'),
    Patient_Count=('Overall Survival (Months)', 'count'),
    ).sort_values(by='Survival_months_mean', ascending = True)

b = a[a['Patient_Count'] > 10]
display(b)
```
